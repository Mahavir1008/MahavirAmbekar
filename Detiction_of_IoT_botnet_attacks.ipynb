```
All_combined_f_28_dummy=pd.get_dummies(All_combined_f_28,columns=['Output'])
DL_Output=All_combined_f_28_dummy.iloc[:,28:]
DL_Input_F=All_combined_f_28_dummy.iloc[:,28:]
```

```
All_combined_f_28_dummy
```

```
DL_Output.shape
```

```
#Using SKLearn
scale=StandardScaler()
scale.fit(DL_Input_F)
scale.transform(DL_Input_F)
DL_Input_F.shape
```

```
Xtrain_Dl,Xtest_Dl,Ytrain_Dl,Ytest_Dl=train_test_split(DL_Input_F,DL_Output,test_size=0.3,random_state=1)
```

```
# Importing the libraries
import numpy as np
import pandas as pd
import tensorflow as tf
```

```
#Xtrain.shape
```

```
# Initializing the ANN
ann = tf.keras.models.Sequential()

# Adding the input layer and the first hidden layer
ann.add(tf.keras.layers.Dense(units=14,activation='relu'))
```

```
# Adding the second hidden layer
ann.add(tf.keras.layers.Dense(units=14, activation='relu'))


# Adding the output layer
ann.add(tf.keras.layers.Dense(units=9, activation='softmax'))
```

```
# Part 3 - Training the ANN

# Compiling the ANN
ann.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```
Ytrain.shape
```

```
Xtrain.shape
```

```
#confusion_matrix(yt,yt)
```

```
# Training the ANN on the Training set
ann.fit(Xtrain_Dl, Ytrain_Dl, batch_size = 500, epochs = 100)
```

```
y_pred = ann.predict_classes(Xtest_Dl)
```
